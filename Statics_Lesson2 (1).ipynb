## **Vocabulary** üî§

**Equilibrium**: The state in which the sum of forces and torques acting on a structure is zero, indicating that the structure is either stationary or moving at a constant velocity.

**Support Forces:** The forces exerted on a structure, such as a highway bridge, at its support points (e.g., foundations, pillars). These forces are essential for maintaining equilibrium.

**Vertical Equilibrium:** The condition in which the sum of vertical forces acting on a structure is zero, ensuring that the structure remains stationary in the vertical direction.

**Horizontal Equilibrium:** The condition in which the sum of horizontal forces acting on a structure is zero, assuming there are no horizontal loads, ensuring the structure remains stationary in the horizontal direction.

**Net Force:** The overall force acting on a structure, calculated as the vector sum of all individual forces acting in different directions.

# **Main Lesson üéí: The First Condition for Equilibrium**

In the realm of mechanics, we examine a key scenario: equilibrium with zero net force and torque. This state signifies that an object, such as a person standing on a frozen pond, or a highway bridge over a lake, is either stationary or its center of mass moves at a constant velocity. Our focus primarily lies in the former, where the bridge or the person remains at rest.

The study of statics enables us to calculate the forces to maintain equilibrium. Statics, a branch of physics, addresses the calculation of forces within objects to prevent significant deformation or fracture.

Even when things appear stationary, like a person standing on a frozen lake, there are multiple forces at play. In this case, gravity pulls the person downward, while the ice below exerts an upward force on the person, counteracting gravity. Because the person remains still, according to Newton's second law, the net force acting on it is zero. This means that the upward force from the ice must be equal in strength to the downward force of gravity on the person. In such a state, the person is considered to be in equilibrium, meaning they are in a state of balance where the forces cancel each other out.

image.png


To ensure that the person remains in a state of "rest", we must apply Newton's second law, which dictates that the sum of all forces acting upon the person must equal zero. These forces are vector quantities, and to maintain equilibrium, it is imperative that the components of the net force are independently zero:

$
\sum F_x = 0
$
$
\sum F_y = 0
$
$
\sum F_z = 0
$

These equations, collectively known as the first condition for equilibrium, are fundamental in evaluating stability.

You've been tasked with evaluating the stability of a person standing on ice on a frozen lake. This assessment involves considering the forces acting on the person and the capacity of the ice to support their weight.

### **Thinking in steps:** ü§î
**Given:**
- Ice Load Capacity = $P_{capacity}$ = $200 \text{lb}$
- Weight = $F_{W}$ = 150 lb

**Step 1**: Calculate Support Forces

To calculate the support forces $S1$ and $S2$ for the person, we can use the equations of equilibrium.

Equilibrium in the vertical direction:
$$
Œ£F_y = 0 = F_{N1} + F_{N2} - F_{W} = 0
$$


But why are $F_{N1}$ and $F_{N2}$ positive and $F_{W}$ is negative?


The choice of signs in the equilibrium equations is a matter of convention and consistency. It's common practice to assign up as positive and down as negative.

> ‚¨Ü positive ‚¨á negative

Based on that convention, the weight of the person acts vertically downward. In the our chosen coordinate system (where upward is positive), the weight is considered negative because it acts in the opposite direction to the positive direction. The support forces from the ice act vertically upward to counteract the weight. In the chosen coordinate system, these support forces are considered positive because they act in the same direction as the positive direction.

The other equation of equilibrium we have is in the horizontal direction (here though we are assuming no horizontal load):
$$
Œ£F_x = 0
$$

In terms of sign convention though, we would have a similar story to what happened above--the choice of signs is a matter of convention and consistency. Commonly people use the right as positive and the left as negative but it does not matter as long as you are consistent and clear in your assumptions when you are writing up your work!

So far we have 2 unknowns and 2 equations but one of them (the horizontal one) doesnt actually give us any information. Above in the problem though we said that the feet would evenly spread the load so our additional constraint here is that $F_{N1} = F_{N2}$.

Now, let's solve these equations:

Using vertical equilibrium, we find that:
$$
F_{N2}= \frac{F_W}{2}
$$


Substituting the given value for the weight of the person $F_W$:

$$
F_{N2} = \frac{150 lb}{2} = 75 \, \text{lb}
$$

Since $F_{N1} = F_{N2}$,
$$
F_{N1} = F_{N2}= 75 lb
$$

**Step 2,** Check if the person is stable:

- If $P_{ice} \leq P_{capacity}$ , the person is stable.
- If $P_{ice} > P_{capacity}$ , the person is not stable.

**Results**

### **Coding time:**  üíª


````python
# Given data
P_capacity = 200  # Ice Bearing Capacity in lb
FW = 150  # Weight in lb

# Step 1: Calculate Support Forces
FN2 = FW / 2  # FN2 is half of the weight
FN1 = FN2  # FN1 is the same as FN2, maintaining symmetry

# Step 2: Check Stability
if FN1 <= P_capacity:  # If FN1 is less than or equal to the ice bearing capacity
    stability_result = "The person is stable."
else:
    stability_result = "The person is not stable."


````

The condition $F_{N1} <= P_{capacity}$ is a comparison, and it evaluates to `True` if $F_{N1}$ is less than or equal to $P_{capacity}$, and `False` otherwise.

If the condition is `True` (i.e., if $v$ is less than or equal to $P_{capacity}$), the code inside the indented block following the if statement is executed.

```

```

This line assigns the string "The person is stable." to the variable `stability_result`.

````python
import matplotlib.pyplot as plt

# Given data
P_capacity = 200  # Ice Bearing Capacity in lb

# List of different weights
weights = [150, 180, 200, 220, 450]  # Modify this list with different weights

# Initialize lists for support forces
support_forces = []

# Calculate support forces for each weight using loops
#This line starts a loop that iterates through each value in the list weights.
#The variable weight takes on each value in the list, one at a time, during each iteration of the loop.
for weight in weights:
    # Inside the loop, this line calculates the support force S2 for the current weight.
    # It assigns half of the current weight to S2.
    # This is based on the assumption that the support forces are symmetric, and S2 is half of the weight.
    FN2 = weight / 2
    # This line assigns the value of S2 to S1. This step is taken to maintain symmetry, assuming equal support forces on both sides.
    FN1 = FN2

    # This line appends the value of S1 to the list support_forces.
    # After each iteration of the loop, the calculated support force for the current weight is added to the list.
    support_forces.append(FN1)

    # This loop goes through each weight in the list, calculates the corresponding support forces S1 and S2,
    # ensures symmetry, and appends the calculated S1 to the list support_forces.
    # The result is a list of support forces corresponding to each weight in the weights list.

# Plotting
plt.plot(weights, support_forces, marker='o', linestyle='-', color='b', label='Support Force (FN1)')
# plt.plot(): This is a function from the matplotlib.pyplot library used to create a line plot.
# weights, support_forces: These are the data points to be plotted. weights is on the x-axis, and support_forces is on the y-axis.
# Each pair (weights[i], support_forces[i]) represents a point on the plot.
# marker='o': This specifies the marker style for each data point. In this case, it uses a circular marker ('o').
# linestyle='-': This specifies the line style between the data points. Here, it uses a solid line ('-').
# color='b': This specifies the color of the line plot. 'b' represents blue.
# label='Support Force (FN1)': This provides a label for the line in the plot. It is useful when you want to add a legend to the plot later. The label is set to "Support Force (FN1)".
# The plt.plot() function is creating a line plot using the weights as x-coordinates, support_forces as y-coordinates, with circular markers, a solid blue line, and a label for the line in the legend.
# The resulting plot will show the relationship between weights and support forces, with each point marked and connected by a blue line.


plt.axhline(y=P_capacity, color='r', linestyle='--', label='Ice Bearing Capacity')  # Red line at FN1 = 200
# plt.axhline(): This is a function from the matplotlib.pyplot library used to add a horizontal line to the plot.
# y=P_capacity: This specifies the y-coordinate at which the horizontal line will be drawn. In this case, it's set to the value of P_capacity.
# color='r': This specifies the color of the horizontal line. 'r' represents red.
# linestyle='--': This specifies the line style for the horizontal line. Here, it uses a dashed line ('--').
# label='Ice Bearing Capacity': This provides a label for the horizontal line in the legend. It's set to "Ice Bearing Capacity".

plt.xlabel('Weight (lb)')  # Label for the x-axis
plt.ylabel('Support Force (FN1) (lb)')  # Label for the y-axis
plt.title('Relationship between Weight and Support Force')  # Title for the plot
plt.legend()  # Adding a legend to explain the plotted line

# Display the plot
plt.show()
````

## **A note about solving for unknowns**

In mathematics, when you have a system of two unknowns, you typically need two non-trivial equations to solve for those unknowns. This concept is rooted in the idea that each equation represents a relationship between the variables, and the solution is the point that satisfies both relationships simultaneously.

Here, you have two equations, and each equation relates our knowns, $F_{N1}$ and $F_{N2}$, in a specific way. Since we didn't have any horizontal forces in the example above, the horizontal equilibrium equation was trivial. We used our knowledge of the problem, that the weight would have to be distributed equally if it was placed in the center to come up with our next equation, $F_{N1} = F_{N2}$. The solution to the system is the values of $F_{N1}$ and $F_{N2}$ that satisfy both equations. You can use various methods, such as substitution or elimination, to find the values of $F_{N1}$ and $F_{N2}$that make both equations true.

The reason you need two equations is that one equation alone doesn't provide enough information to uniquely determine the values of both unknowns. Two equations create an overlap or intersection point in the solution space, narrowing down the possibilities to a specific solution.

If you have more unknowns, you would generally need as many equations as the number of unknowns to uniquely determine the values of those unknowns. This is a fundamental concept in solving systems of linear equations, and it extends to more complex mathematical problems as well.
