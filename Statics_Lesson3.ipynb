## **Sticking points** ‚ö†

**Sign Convention for Torque:** The introduction of a negative sign in the torque equation and its association with clockwise and counterclockwise rotations might be confusing. Ensuring that students understand the chosen sign convention and its consistent application is crucial.

**Torque vs moment:** "Torque" and "moment" are terms often used interchangeably. When we talk about our condition for equilibrium we use torque since it is the more generic term about the tendency of a force to raotate an object about an axis, We will use moment when we are referring to a specific external moment connection or a value we are calculating at a specific point.

**Moment Connections and Additional Torques:** The idea of moment connections and the inclusion of additional torques beyond the direct weight force might be a potentially complex concept. Students may struggle to understand why certain torques exist and why they are essential for rotational equilibrium.

## **Vocabulary** üî§

**Torque** is a measure of the rotational force or tendency to rotate applied to an object. It is calculated as the product of the lever arm distance and the force applied perpendicular to the lever arm.

**Moment connection:** A moment connection refers to a joint or connection in a structure that allows for the transmission of both forces and torques between connected components. It is significant in providing stability and preventing rotation in structural systems.

**Lever arm:** The lever arm, also known as the moment arm, is the perpendicular distance from the pivot point to the line of action of a force. It is a crucial parameter in torque calculations.

**Stable equilibrium:** Stable equilibrium occurs when an object, when disturbed, returns to its original position. For stable equilibrium, the object's center of gravity should be directly above its point of support.

**Unstable equilibrium**: Unstable equilibrium occurs when a slight disturbance leads to continuous motion, and the object does not return to its original position. This often happens when the center of gravity shifts away from the point of support.

**Material fatigue**: Material fatigue is the weakening of a material over time due to repeated loading and unloading. It can lead to structural failure and is a consideration in setting safe operating limits.

# **Main Lesson üéí: The Second Condition For Equilbrium**

image.png

While those equations we looked at last lesson are necessary for an object to be in equilibrium, they do not always suffice. In the figure above, you have a pickup truck equipped with a folding truck-mounted crane with an electric winch and you need to lifting fish from the ground. Despite the fact that the forces $F_W$ and $F_N$ would be in equilbrium, they generate a net torque capable of rotation. Thus for that crane to not rotate, there has to be another equation at play!

$$
\sum\tau=0
$$
Therefore we can establish the second condition for equilibrium: the sum of torques acting on an object, calculated about any axis, must equate to zero.

But what is this torque, $\tau$?

$$
\tau= d_{arm} \times F_{applied}
$$

$d_{arm}$ is the lever arm or the perpendicular distance from the pivot point to the line of action of the force.

$F_{applied}$ is the applied force or in this case the load of fish.

Let's try calculating what the torque at point A, the pivot point, would be in an example.

## **Example 1**


In the figure above, you have a pickup truck equipped with a folding truck-mounted crane with an electric winch and you need to lift 1000lb of fish from the ground. The crane is attached to the bed of the truck and the electronic winch is used to lift the load vertically. The length of the crane arm from the pivot point to the hook is 46 inches. Calculate how much torque exists at the pivot point when the fish are hoisted up.

### **Thinking in steps:** ü§î


**Given:**
- Weight = $F_W$ = 1000 lb
- Arm distance = $d_{arm}$ = 46" (about 3.8 ft)

**Step 1:** Write out your equations of equilbrium for forces like we did in the previous lesson

Equilibrium in the vertical direction:
$$
Œ£F_y = 0 = F_N - F_W = 0
$$

Equilibrium in the horizontal direction is trivial (assuming no horizontal load):
$$
Œ£F_x = 0
$$

**Step 2:** Now we also have another equation to consider: torque!

$$
\sum\tau_{A}=0 = d_{arm} \times F_{W} + d_{arm} \times F_{N}
$$

Now we are intereseted in understanding how much torque occurs at point A. So how far away is point A from where our normal force is applied?

They are coincident! That means that $d_{arm}$ for our $F_{N}$ is 0 and so that term goes to 0 for our equation above. That means we are just left with:

$$
\sum\tau_{A}=0 = d_{arm} \times F_{W}
$$

So now we have three equations that we can use to understand how much torque exists at the pivot point when the fish are hoisted up.

$$
\sum\tau_{A}=0 = 3.8\text{ft} \times 1000 \text{lb}
$$

But that doesn't make sense!? How does 0 = 380 lbft?

Well it turns out that there is another term in that figure that we did not account for in our torque equation, the actual torque we are trying to solve for! So instead our equation should look like

$$
\sum\tau_{A}=0 = 3.8\text{ft} \times 1000 \text{lb} - \tau_A
$$

**Results:**

After substituting our real values back into that equation we get

$$
\tau_{A} = 380 \text{lbft in the clockwise direction}
$$

3.  How did we know what direction to draw it in our freebody diagram?
4.  And why is there a negative sign in front of $\tau_A$? And what does "in the clockwise direction" mean?

### **Clarifying confusing points:**‚ùì‚ùì‚ùì


So let's start with one of those questions at a time:




**1. Where did that term come from?**

In the context of a crane lifting a load, the torque $\tau_A$ at point A represents the contribution to the rotational equilibrium of the system due to factors other than the weight of the fish. The equation $\sum\tau = 0$ is a statement that the sum of all torques acting on the system must be zero for the system to be in rotational equilibrium.

Now, consider the corrected equation:

$$
\sum\tau_{A} = 0 = d_{\text{arm}} \times F_{W} - \tau_A
$$

The first term $d_{\text{arm}} \times F_{W}$ accounts for the torque generated by the weight of the fish acting at a distance $d_{\text{arm}}$ from the pivot point. However, there might be other torques $\tau_A$ acting at point A like the components within the crane, such as motors and the fastening of the crane to the truck bed (can be referred to as a moment connection), which create torques that affect the equilibrium.

When we refer to the "fastening of the crane to the truck bed" as a "moment connection," we are describing a type of joint or connection that allows for the transmission of both forces and torques between the crane and the truck bed. A moment connection allows the transmission of torques or moments between the crane and the truck bed. This torque transmission is significant when considering the rotational equilibrium of the system.

In essence, $\tau_A$ is a placeholder for additional torques that may exist due to various factors in the system. It's the *sum* of all torques at the pivot point A that prevent the crane from tipping over. If we didn't fasten the crane to the truck and instead just placed it in the bed, our system would not be able to generate $\tau_A$ and it would fall over! Including $\tau_A$ in the equation ensures that the rotational equilibrium condition is comprehensive, accounting for all relevant torques in the system.



**2. Why do we have $\tau_A$ on one end of the crane but not the other?**

$\tau_A$ exists because of the moment connection that we just talked about. It doesnt exist on both sides of the crane, it just exists where the bolts are fastened to the truck bed. Since there isn't a similar mechanism on the other end of our crane (where our fish are hooked on) that side doesn't have the capacity to counter act rotation.



**3. How did we know what direction to draw it?**

Okay well let's think about the situation above before we knew $\tau_A$ existed. If we didn't fasten the crane to the truck bed and we tried to hoist our fish up, the crane would tip clockwise about the pivot point.

But since we did fasten our crane to the truck bed, when we go to hoist our fish up, the moment connection is trying to prevent the crane from rotating clockwise and so it is acting counterclockwise.



**4. Why is there a negative sign in front of $\tau_A$?**

Again, it is important to remember that the choice of signs in the equilibrium equations is a matter of convention and consistency. When looking at rotations it's common practice to assign clockwise as positive and counterclockwise as negative. Based on that convention, the weight of the fish would spin the crane clockwise around point A and it is positive; $\tau_A$ though is acting in the opposite direction and so we write that with a negative sign.

So looking back at question 3, what would happen if we had drawn $\tau_A$ going the othe way and we had assumed that $\tau_a$ was trying to rotate the crane clockwise about point A? In that case we would have just ended up with a negative value which would have indicated that we chose our direction wrong.

The recommended maximum load or torque limits provided by manufacturers are typically based on a combination of factors.
* Manufacturers often incorporate safety margins into their specifications to ensure the safe operation of the equipment. The recommended limits are set lower than the theoretical maximum capacity to provide a buffer for uncertainties, variations in materials, potential wear and tear, and other unforeseen factors.
* Materials, especially metals, can experience fatigue and wear over time with repeated loading and unloading. Manufacturers consider the material properties and design the equipment to withstand a certain number of loading cycles without failure. Setting conservative load limits helps extend the life of the equipment and reduces the risk of material fatigue.

While it might seem that the equipment could handle a higher load based on theoretical calculations, the recommended limits are established to prioritize safety and reliability. ## **A note about stability and balance**

An object in static equilibrium, when undisturbed, experiences no translational or rotational acceleration because the total sum of forces and torques acting on it is zero. Generally, an object with its center of gravity (cg) situated directly above its point of support, such as a net full of fish, tends to be in stable equilibrium. Stable equilibrium means that if the net is gently nudged to one side and then released, it will return to its original position.



image.png

On the other hand, if the net is delicately displaced, and its center of mass shifts slightly away from the geometric center, it may start sliding further in that direction. This represents unstable equilibrium, where a slight disturbance leads to continuous motion. If the net continues to tip past the second picture so that the center of gravity has past the pivot point, $F_N$, we get to the third picture where the net is toppling. This is because the normal force upward on the object can be exerted only within the area of contact. That means that if the weight acts beyond this area, a net torque will topple the fish.
