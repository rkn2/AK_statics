# **What is a code outline?**



A code outline is akin to a blueprint or a structured plan that serves as a guide for solving a specific problem. It is a high-level representation of the logical steps and key components required to achieve a desired outcome, all before diving into the nitty-gritty details of writing the actual code. This preliminary structure allows students learning to code, to organize their thoughts, identify patterns, and establish a clear roadmap for their solution.

Now, let's break down the key components of a code outline:

**Step 1: Problem Understanding:** Clearly articulate the problem at hand. What is the physics scenario you're trying to model or solve? Understand the problem thoroughly before attempting to convert it into code.

**Step 3: Variable Assignments:** Identify the constants and save them as variables in your code. These are the entities whose values will be constant as the program runs and we will need to call on them. Assign initial values to these variables based on the problem's context.

**Step 4: Calculations:** Outline the mathematical or logical calculations required to solve the problem. This could involve applying physics formulas, performing arithmetic operations, or implementing algorithms relevant to the scenario.

**Step 5: Print Statements:** Define where and how the results should be displayed. This involves using print statements or equivalent output mechanisms to communicate the computed values or outcomes.

Plot the trajectory of your object using matplotlib with the parametric equations for projectile motion.

````python
# Clearly define the physics problem and identify what needs to be calculated: trajectory.
# Think about what the governing equations are for your problem at hand and write them down so you can refer to them when you ar
# For projectile motion, we will be using the parametric equations:
  # x(t) = v_0x * t
  # y(t) = h_0 + v_0y * t + 0.5 * a * t^2
    # Where:
    # - v_0x is the x component of the initial velocity
    # - v_0y is the y component of the initial velocity
    # - h_0 is the initial height
    # - a is the constant acceleration (gravity, assumed constant)

# Define your given data
# Set a equal to -32 ft/s^2 (acceleration due to gravity)
# Set v_0 equal to the initial velocity
# Set theta equal to the launch angle in degrees
# Set h_0 equal to the initial height (let's assume it's 0 for simplicity)

# Define your formulas that you will need
# To calculate the x and y positions, use the parametric equations for projectile motion

# Plot the trajectory using matplotlib
````

and here is your code inside your outline

````python
# - Clearly define the physics problem and identify what needs to be calculated: trajectory.
# - Think about what the governing equations are for your problem at hand and write them down so you can refer to them when you a
# - For projectile motion, we will be using the parametric equations:
#   x(t) = v_0x * t
#   y(t) = h_0 + v_0y * t + 0.5 * a * t^2
#     Where:
#     - v_0x is the x component of the initial velocity
#     - v_0y is the y component of the initial velocity
#     - h_0 is the initial height
#     - a is the constant acceleration (gravity, assumed constant)

# - Define your given data
# - Set a equal to -32 ft/s^2
# - Set v_0 equal to the initial velocity
# - Set theta equal to the launch angle in degrees
# - Set h_0 equal to the initial height
a = -32  # acceleration due to gravity in ft/s^2
v0 = 10  # initial velocity in ft/s
theta = 45  # launch angle in degrees
h0 = 0  # initial height

from math import sin
from math import cos

# - Define your formulas that you will need
# - To calculate the x and y positions, use the parametric equations for projectile motion
v0x = v0 * cos(theta)
v0y = v0 * sin(theta)

# - Plot the trajectory using matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Time points
t_points = np.linspace(0, 2 * v0y / -a, num=100)
# x and y positions
x_points = v0x * t_points
y_points = h0 + v0y * t_points + 0.5 * a * t_points**2

# Plotting
plt.plot(x_points, y_points)
plt.title('Projectile Motion Trajectory')
plt.xlabel('Horizontal Distance (ft)')
plt.ylabel('Vertical Distance (ft)')
plt.grid(True)
plt.show()
````

# **Sticking points:**

Creating a code outline can be a crucial step in helping students understand the logic and structure of a program! However, there are some. common sticking points you could run into when making a code outline. Here are some of those points and how to mitigate them:


**Unclear Variable Definitions:**
  * Sticking Point: Students may struggle with understanding the difference in variables or constants used in the problem versus their code.
  * Mitigation: Encourage students to use clear and explicit comments in their code. For instance, when a constant like gravity `g` is declared in the code, students should add comments clarifying that, although it's stored as a variable, its value remains constant throughout the problem.



```
# Constants
g = 9.8  # Acceleration due to gravity in ft/s^2 (remains constant)
```



**Complexity Overload:**
  * Sticking Point: Teachers may unintentionally make the code outline too complex, leading to confusion.
  * Mitigation: Break down the problem into smaller, manageable steps. Clearly define the tasks and guide students through one step at a time. You can even break code into multiple code blocks if that makes it easier for the students to read.
