## **Vocabulary** üî§

**Piers:** Vertical structures that provide support to the bridge deck. Positioned beneath key points along the bridge span, piers serve as strong, vertical supports that counteract the downward force of the bridge's weight and other loads.

**Abutments:** The external supports of a bridge, typically located at its ends. Abutments transfer the bridge's vertical loads to the ground and provide lateral support to resist horizontal forces.

**Internal forces:** An internal force is a force that develops within a structure in response to external loads or changes in support conditions. These forces arise to maintain the equilibrium of the structure and ensure that it remains stable and structurally sound. Internal forces include shear forces, normal forces, and moments, and they act within the structural elements such as beams, columns, and trusses. These forces are distributed throughout the structure and are crucial for maintaining balance and stability. Understanding internal forces is fundamental to analyzing and designing structures in the field of structural engineering.

## **Example**

In our first analysis of the bridge's static equilibrium, let's consider a simplified version where we ignore the presence of the piers and only consider forces from the abutments ($F_{A1}$ and $F_{A2}$ ) and the weight of the bridge ($F_{W}$ ).

image.png

So let's estimate that the bridge is 300 ft long and weighs 600,000 lb.

If you are curious about how we got the weight of the bridge, we looked up what is the density of a concrete highway bridge deck and found it to be approximately 100 lb / ft2. If the highway bridge is 20' wide and 300' long, then the bridge would be
$$
20 \text{ ft} \times 300 \text{ ft} \times 100 \frac{lb}{ft^2} ‚âà 600,000 lb
$$


So your task is to figure out how much each support is carrying for the bridge. So in this example, there are two unknowns $F_{A1}$ and $F_{A2}$ which means we need two equations to solve.



There are two ways to solve this problem:

1. We can assume that $F_{A1} = F_{A2}$ like we did in an earlier lesson and use this as our second equation.

2. We can solve this in a more general way where we do not just assume $F_{A1} = F_{A2}$ and need to use the rotational equilibrium condition as well.

Let's go with option 2 for practice with the concepts. You can always check yourself by doing this problem again with option 1 and comparing your answers!

### **Thinking in steps:** ü§î

**Step 1:** Write out your equations of equilbrium for forces like we did in the previous lesson

So let's assume that $FW$ is acting at the center of the bridge and that the abutments are 300' apart from each other.  In terms of equations we have:

Equilibrium in the vertical direction:
$$
Œ£Fy = 0 = F_{A1} + F_{A2} - F_W = 0
$$

$$
Œ£Fy = 0 = F_{A1} + F_{A2} - 600,000\text{lb} = 0
$$

Equilibrium in the horizontal direction (assuming no horizontal load):
$$
Œ£Fx = 0
$$

**Step 2:** Now we also have another equation to consider: torque!

$$
\sum(\tau_{A1})=0 = d_{arm} \times F_{A1} - d_{arm} \times F_{A2} + d_{arm} \times F_{W}
$$

Remember that the term associated with the weight will be trying to turn the bridge clockwise about point 1 and so we consider that to be postive. Additionally $d_{arm}$ for $F_{A1}$ is 0 since there is no distance between point 1 and where that force is applied.

**Results:**

$$
\sum(\tau_{A1})=0 = -300\text{ft} \times F_{A2} + \frac{300}{2} \text{ft} \times 600,000 \text{lb}
$$

Solving for our unknowns we get
$$
F_{A2} = 300,000 \text{lb}
$$
$$
F_{A1} = 300,000 \text{lb}
$$

```python
# This code cell can be deleted so the students have to write it on their own from scratch
# This code cell can be modified to be pseudo code so students have to complete the code on their own
      # Here is an example of what you could give for a code outline
      # Import necessary library
      # Given values
      # Set distance_to_second_abutment to 300
      # Set total_weight to 1000000
      # Using range to create a list of distances to the weight (from 0 to 300 feet)
      # Set distances_to_weight using range
      # Calculate corresponding forces for each distance using loops
      # Initialize empty lists forces_A1 and forces_A2
      # For each distance in distances_to_weight:
      #     Calculate force_A1: total_weight - (distance * (total_weight / distance_to_second_abutment))
      #     Calculate force_A2: distance * (total_weight / distance_to_second_abutment)
      #     Append force_A1 to forces_A1
      #     Append force_A2 to forces_A2
      # Plotting
      # Plot forces_A1 and forces_A2 against distances_to_weight
      # Add labels and title to the plot
      # Display a legend to identify each line
      # Show the plot
# This code cell can be left as is so students have an example to work from and they can be encouraged to just change values

import matplotlib.pyplot as plt

# Given values
distance_to_second_abutment = 300  # Distance between abutments in feet
total_weight = 600000  # Total weight of the bridge in pounds

# Using range to create a list of distances to the weight (from 0 to 300 feet)
# The third argument is the step size (increments), here set to 3 for simplicity
# Adjust the step size based on your desired level of detail
distances_to_weight = range(0, distance_to_second_abutment + 1, 3)

# Calculate corresponding forces for each distance using loops
forces_A1 = []
forces_A2 = []

for distance in distances_to_weight:
    force_A1 = total_weight - (distance * (total_weight / distance_to_second_abutment))
    force_A2 = distance * (total_weight / distance_to_second_abutment)
    forces_A1.append(force_A1)
    forces_A2.append(force_A2)

# Plotting
plt.plot(distances_to_weight, forces_A1, label='FA1')
plt.plot(distances_to_weight, forces_A2, label='FA2')

# Adding labels and title
plt.xlabel('Distance to Weight from First Abutment (feet)')
plt.ylabel('Force (lb)')
plt.title('Effect of Weight Position on Support Forces')
plt.legend()

# Show the plot
plt.show()
```

And that makes sense! What it is saying is that when we put all the load directly onto the first abutment, $F_{A1}$ bears all the load. As we move the load onto the bridge deck though, we start to split it with $F_{A2}$. Then as we move the load all the way across to the second abutment,$F_{A2}$ bears all the load.

It turns out though that our bridge is a bit more complicated than what we looked at above and we do actually have piers in the middle. These piers serve a dual purpose: not only do they minimize the load borne by the supports, but they also play a crucial role in reducing internal forces within the bridge structure.

To gain insight into the internal forces at play, imagine making a cut at three-quarters of the span and examining only the left portion of the bridge. At the cut end of this section, we must account for the internal upward force (referred to as the shear force), the internal horizontal force (referred to as the normal force, though it might be trivial in this context), and the internal moment at the cut point.

It turns out, in structural mechanics that internal moment, $M_{cut}$, can only be so large. The materials used in bridges, such as concrete or steel, have specific strength limits. Beyond a certain point, these materials may experience deformation, yielding, or failure. The internal moment must be kept within the range that ensures the structural integrity of the materials.

So let's assume that $M_{allowable} = 15,000,000 \text{lb ft}$. Calculate what $M_{cut}$ would be for the example above.

So let's assume that $FW$ is again acting at the center of the bridge and that the abutments are 300' apart from each other and we made our cut 225' from abutment 1.  In terms of equations we have:

Equilibrium in the vertical direction:
$$
Œ£Fy = 0 = F_{A1} + F_{Cut} - F_W = 0
$$

So we can substitute in what we know from our earlier problem. The support reactions 'don't care' that we are looking at this internal section of our bridge. They are going to be the same as long as the external loads remain the same, which they have. So we can substitute in for $F_{A1} = 300,000 \text{lb}$

$$
Œ£Fy = 0 = 300,000\text{lb} + F_{Cut} - 600,000\text{lb} = 0
$$

Equilibrium in the horizontal direction (assuming no horizontal load):
$$
Œ£Fx = 0
$$

And we also have a third equation

$$
\sum\tau_{Cut}=0 = d_{arm} \times F_{A1} - d_{arm} \times F_{W} - d_{arm} \times F_{Cut} + M_{Cut}
$$


$$
\sum\tau_{Cut}=0 = (225 \text{ft}) \times 300,000 \text{lb} - (225\text{ft}-150\text{ft}) \times 600,000 \text{lb} - 0 \times F_{Cut} + M_{Cut}
$$

which simplifies to

$$
\sum\tau_{Cut}=0 = (225 \text{ft}) \times 300,000 \text{lb} - (225\text{ft}-150\text{ft}) \times 600,000 \text{lb} + M_{Cut}
$$

So now we can use our two equations and solve for $F_{cut}$ and $M_{Cut}$.

$$
F_{cut} = 300,000 \text{lb}
$$

$$
M_{Cut} = - 22,500,000 \text{lb ft}
$$

Since our moment came out negative that means we originally drew it acting in the wrong direction. In reality our internal moment is trying to "bend the beam back up" at the cut. If we had originally drawn that moment as counter clockwise on the left cut of our beam, then we would have had this come out positive. Can you write out how you would solve this pythonically?

### **Coding time:**  üíª

```python
# Given values
FW_value = 600000  # lb
FA1_value = FW_value /2  # lb
distance_arm = 225  # ft
distance_from_cut_to_FW = 150  # ft

# Solve for FCut
FCut_value = FW_value - FA1_value

# Solve for MCut
MCut_value = distance_arm * FA1_value - (distance_arm - distance_from_cut_to_FW) * FW_value

# Display the results
print("FCut:", FCut_value, "lb")
print("MCut:", MCut_value, "lb ft")
```
