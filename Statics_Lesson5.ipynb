## **Vocabulary** 🔤

**Piers:** Vertical structures that provide support to the bridge deck. Positioned beneath key points along the bridge span, piers serve as strong, vertical supports that counteract the downward force of the bridge's weight and other loads.

**Abutments:** The external supports of a bridge, typically located at its ends. Abutments transfer the bridge's vertical loads to the ground and provide lateral support to resist horizontal forces.

**Internal forces:** An internal force is a force that develops within a structure in response to external loads or changes in support conditions. These forces arise to maintain the equilibrium of the structure and ensure that it remains stable and structurally sound. Internal forces include shear forces, normal forces, and moments, and they act within the structural elements such as beams, columns, and trusses. These forces are distributed throughout the structure and are crucial for maintaining balance and stability. Understanding internal forces is fundamental to analyzing and designing structures in the field of structural engineering.

## **Understanding Cut Sections: A Physical Analogy** 🪚

### **Imagine Sawing a Wooden Plank**

The "cut section method" might sound abstract, but it's actually quite intuitive!

**Physical Analogy:**

1. **Start with a plank** holding up a weight
   - The plank is in equilibrium (not moving)
   - Internal forces hold it together

2. **Imagine sawing through the plank** at some point
   - Now you have two pieces
   - What forces would you need to apply to keep each piece in place?

3. **Those forces reveal the internal forces!**
   - The force you'd need to push/pull = **Shear force**
   - The twisting you'd need to apply = **Bending moment**

### **Why This Works:**

Before the cut:
- Internal forces in the material hold everything together
- These forces are "hidden" inside the structure

After the imaginary cut:
- We expose these internal forces
- We can calculate them using equilibrium equations

### **Key Insight:**

💡 The "cut section method" doesn't actually cut anything - it's a mental tool that lets us analyze internal forces by imagining what would happen if we did cut the structure.

**Think of it like:**
- Cutting a rope to see the tension
- Sawing a beam to see the shear and moment
- Opening a pressure vessel to see the internal pressure


## **Vocabulary Recap** 📚

Let's review and reinforce key terms:

### **Shear Force**
Forces that try to slide one part of a structure past another part, like scissors cutting paper.
- **Symbol:** V
- **Units:** lbs or N
- **Example:** When you step on a diving board, shear forces try to "cut" through it

### **Bending Moment (Moment)**
The tendency of a force to cause rotation or bending.
- **Symbol:** M
- **Units:** lb·ft or N·m
- **Example:** When you push on a door handle, you create a moment about the hinges

### **Cut Section**
An imaginary "cut" through a structure that exposes internal forces.
- **Purpose:** Reveals hidden internal forces
- **Method:** Apply equilibrium to the exposed section
- **Analogy:** Like sawing a plank to see what holds it together

### **Equilibrium**
A state where all forces and moments balance out (sum to zero).
- **Force equilibrium:** ΣF = 0
- **Moment equilibrium:** ΣM = 0
- **Result:** No acceleration, no rotation

### **Abutment**
A support structure at the end of a bridge.
- **Function:** Transfers bridge loads to the ground
- **Location:** At each end of the bridge

### **Pier**
An intermediate support structure between bridge ends.
- **Function:** Supports the bridge from below
- **Location:** Between abutments (for longer bridges)

## **Example**

In our first analysis of the bridge's static equilibrium, let's consider a simplified version where we ignore the presence of the piers and only consider forces from the abutments ($F_{A1}$ and $F_{A2}$ ) and the weight of the bridge ($F_{W}$ ).

So let's estimate that the bridge is 300 ft long and weighs 600,000 lb.

If you are curious about how we got the weight of the bridge, we looked up what is the density of a concrete highway bridge deck and found it to be approximately 100 lb / ft2. If the highway bridge is 20' wide and 300' long, then the bridge would be
$$
20 \text{ ft} \times 300 \text{ ft} \times 100 \frac{lb}{ft^2} ≈ 600,000 lb
$$


So your task is to figure out how much each support is carrying for the bridge. So in this example, there are two unknowns $F_{A1}$ and $F_{A2}$ which means we need two equations to solve.



There are two ways to solve this problem:

1. We can assume that $F_{A1} = F_{A2}$ like we did in an earlier lesson and use this as our second equation.

2. We can solve this in a more general way where we do not just assume $F_{A1} = F_{A2}$ and need to use the rotational equilibrium condition as well.

Let's go with option 2 for practice with the concepts. You can always check yourself by doing this problem again with option 1 and comparing your answers!

### **Thinking in steps:** 🤔

**Step 1:** Write out your equations of equilbrium for forces like we did in the previous lesson

So let's assume that $FW$ is acting at the center of the bridge and that the abutments are 300' apart from each other.  In terms of equations we have:

Equilibrium in the vertical direction:
$$
ΣFy = 0 = F_{A1} + F_{A2} - F_W = 0
$$

$$
ΣFy = 0 = F_{A1} + F_{A2} - 600,000\text{lb} = 0
$$

Equilibrium in the horizontal direction (assuming no horizontal load):
$$
ΣFx = 0
$$

**Step 2:** Now we also have another equation to consider: torque!

$$
\sum(\tau_{A1})=0 = d_{arm} \times F_{A1} - d_{arm} \times F_{A2} + d_{arm} \times F_{W}
$$

Remember that the term associated with the weight will be trying to turn the bridge clockwise about point 1 and so we consider that to be postive. Additionally $d_{arm}$ for $F_{A1}$ is 0 since there is no distance between point 1 and where that force is applied.

**Results:**

$$
\sum(\tau_{A1})=0 = -300\text{ft} \times F_{A2} + \frac{300}{2} \text{ft} \times 600,000 \text{lb}
$$

Solving for our unknowns we get
$$
F_{A2} = 300,000 \text{lb}
$$
$$
F_{A1} = 300,000 \text{lb}
$$

## **Reusable Functions for Bridge Analysis** 🔧

Instead of hard-coding values, let's create functions we can reuse!


```python
def calculate_support_reactions(total_load, bridge_length, load_position):
    """
    Calculate support reactions for a simply supported bridge.
    
    Parameters:
    - total_load: Total load on bridge (lbs)
    - bridge_length: Total bridge length (ft)
    - load_position: Distance from left abutment to load (ft)
    
    Returns:
    - R_A: Reaction at left abutment (lbs)
    - R_B: Reaction at right abutment (lbs)
    """
    # Input validation
    if total_load < 0:
        raise ValueError("Load cannot be negative")
    if load_position < 0 or load_position > bridge_length:
        raise ValueError(f"Load position must be between 0 and {bridge_length} ft")
    
    # Calculate reactions using moment equilibrium
    R_B = (total_load * load_position) / bridge_length
    R_A = total_load - R_B
    
    return R_A, R_B

def calculate_shear_at_point(R_A, load, x, load_position):
    """
    Calculate shear force at a specific point along the bridge.
    
    Parameters:
    - R_A: Reaction at left support (lbs)
    - load: Applied load (lbs)
    - x: Position where we want shear (ft from left)
    - load_position: Position of applied load (ft from left)
    
    Returns:
    - V: Shear force at position x (lbs)
    """
    if x < load_position:
        # Before the load
        V = R_A
    else:
        # After the load
        V = R_A - load
    
    return V

def calculate_moment_at_point(R_A, load, x, load_position):
    """
    Calculate bending moment at a specific point.
    
    Parameters:
    - R_A: Reaction at left support (lbs)
    - load: Applied load (lbs)
    - x: Position where we want moment (ft from left)
    - load_position: Position of applied load (ft from left)
    
    Returns:
    - M: Bending moment at position x (lb·ft)
    """
    if x < load_position:
        # Before the load
        M = R_A * x
    else:
        # After the load
        M = R_A * x - load * (x - load_position)
    
    return M

# Example usage
print("="*60)
print("BRIDGE ANALYSIS EXAMPLE")
print("="*60)

# Bridge parameters
bridge_length = 300  # ft
truck_weight = 600000  # lbs
truck_position = 150  # ft (middle of bridge)

# Calculate support reactions
R_A, R_B = calculate_support_reactions(truck_weight, bridge_length, truck_position)

print(f"\nBridge: {bridge_length} ft long")
print(f"Truck: {truck_weight:,} lbs at {truck_position} ft from left")
print(f"\nSupport Reactions:")
print(f"  Left abutment (R_A): {R_A:,.0f} lbs")
print(f"  Right abutment (R_B): {R_B:,.0f} lbs")
print(f"\nVerification: R_A + R_B = {R_A + R_B:,.0f} lbs")

# Calculate shear and moment at midpoint
x = 150  # ft
V = calculate_shear_at_point(R_A, truck_weight, x, truck_position)
M = calculate_moment_at_point(R_A, truck_weight, x, truck_position)

print(f"\nAt x = {x} ft:")
print(f"  Shear force: {V:,.0f} lbs")
print(f"  Bending moment: {M:,.0f} lb·ft")
print("="*60)
```

```python
# Interactive Widgets for Bridge Analysis
# ========================================

try:
    from ipywidgets import interact, FloatSlider
    import matplotlib.pyplot as plt
    import numpy as np
    
    def interactive_bridge_analysis(load, position, length):
        """
        Interactive bridge analysis with real-time updates.
        """
        # Calculate reactions
        R_A, R_B = calculate_support_reactions(load, length, position)
        
        # Create position array
        x = np.linspace(0, length, 100)
        
        # Calculate shear and moment along bridge
        shear = [calculate_shear_at_point(R_A, load, xi, position) for xi in x]
        moment = [calculate_moment_at_point(R_A, load, xi, position) for xi in x]
        
        # Create plots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
        
        # Shear diagram
        ax1.plot(x, shear, 'b-', linewidth=2)
        ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
        ax1.grid(True, alpha=0.3)
        ax1.set_ylabel('Shear Force (lbs)', fontsize=12)
        ax1.set_title('Shear Force Diagram', fontsize=14, fontweight='bold')
        ax1.axvline(x=position, color='r', linestyle='--', label='Load position')
        ax1.legend()
        
        # Moment diagram
        ax2.plot(x, moment, 'r-', linewidth=2)
        ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
        ax2.grid(True, alpha=0.3)
        ax2.set_xlabel('Position along bridge (ft)', fontsize=12)
        ax2.set_ylabel('Bending Moment (lb·ft)', fontsize=12)
        ax2.set_title('Bending Moment Diagram', fontsize=14, fontweight='bold')
        ax2.axvline(x=position, color='r', linestyle='--', label='Load position')
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
        
        # Print results
        print(f"Support Reactions: R_A = {R_A:,.0f} lbs, R_B = {R_B:,.0f} lbs")
        print(f"Max Moment: {max(moment):,.0f} lb·ft at x = {x[moment.index(max(moment))]:.1f} ft")
    
    # Create interactive widget
    interact(
        interactive_bridge_analysis,
        load=FloatSlider(min=100000, max=1000000, step=50000, value=600000, description='Load (lbs)'),
        position=FloatSlider(min=0, max=300, step=10, value=150, description='Position (ft)'),
        length=FloatSlider(min=100, max=500, step=50, value=300, description='Length (ft)')
    )
    
except ImportError:
    print("⚠️ ipywidgets not installed. Run: pip install ipywidgets")
    print("Then restart the kernel and try again.")
```
