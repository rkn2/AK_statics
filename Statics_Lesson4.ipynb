# **Main Lesson üéí**

The study of statics holds significant importance as it provides a framework for determining specific forces acting upon a structure, particularly when certain forces acting on or within the structure are already known. In the context of statics, we primarily examine scenarios in which all forces operate within a single plane. This allows us to formulate two force equations, one for the x-component and another for the y-component, as well as a torque equation, resulting in a total of three fundamental equations. It is worth noting that not all three equations are always necessary for a given problem; their applicability depends on the specific situation at hand.

Like, if the person is walking on a frozen river in a straight line without any significant tilting or rotation, we may only need the two force equations (x and y components) to analyze the forces acting on them; the person is not actively trying to rotate or pivot while walking. Therefore, the torque equation, which deals with rotational forces, may not be relevant.

Let's conceptually look at another example together and walk through the general steps of how to handle a statics problem. Imagine you are driving a truck with a trailer full of toys (snow mobiles, etc) across a frozen river.

## **Formalizing our thinking in steps** üß†
**Step 1: Identify the Object and Its Forces**

* Choose one object you want to study at a time.
  * So for our example, let's look at the trailer on that truck. Ignore the front of the truck itself for this example and just focus on the wheel carrying the load down from the trailer to the ice.

* Create a clear diagram that shows all the forces acting on the chosen object.
  * In our case we would have the gravity of the trailer and the supports from the ice at A and B.
* Indicate where these forces apply on the object.
  * Gravity would act downward at the center of mass and the normal force would act at each of the supports to counter that so let's draw them going upward.

If you're unsure about the direction of a force, just make a choice. If you picked the wrong direction, your calculation will have a negative sign in the result!

image.png

**Step 2: Choose a Coordinate System**

* Select a convenient coordinate system (like x and y axes) and break down the forces into their components along these axes.

  * So in our example, we only have forces acting in one direction so we only need to look at what is happening in the y axis.

**Step 3: Write Equilibrium Equations**

* Use letters to represent unknown quantities.
  * So here we would have $F_A$ for the reaction at $A$, $F_B$ for the reaction at $B$, and $F_W$ for the weight of the trailer.
* Write two equilibrium equations for the forces: $Œ£F_x = 0$ and $Œ£F_y = 0$.
  * In our case we don't have any forces acting in our x direction but we do have three in our y direction that we should include in that formula.

  $$
Œ£Fx = 0
$$
$$
Œ£Fy = 0 = F_A + F_B - F_W
  $$

* To consider rotational forces (torques), write a torque equation: $Œ£œÑ_{pt} = 0$ about a specific point. You can choose a convenient axis for writing the torque equation such as one of your unknowns.  
  *  In our case lets look at the torque about point A.
$$
Œ£œÑ_A = 0 =  F_A \times d_1 - F_W \times d_2 + F_B \times d_3
$$
  * Pay attention to correctly determining the lever arm (the distance from the axis of rotation) for each force. Assign a plus (+) or minus (-) sign to indicate the direction of rotation (counterclockwise or clockwise).
    * In our case when we are looking about point A, $F_A$ has a moment arm of 0 so that will not cause our trailer to spin clockwise or counterclockwise and is neither negative or positive.
    * $F_W$ will cause the trailer to spin clockwise so we assign that a "-".
    * $F_B$ will cause the trailer to spin counterclockwise so we assign that a "+".






**Step 4: Solve the Equations**
* Solve these equations to find the unknowns.
* Keep in mind that with three equations ($Œ£Fx = 0$, $Œ£Fy = 0$, and $Œ£œÑ_{pt} = 0$), you can solve for a maximum of three unknowns.
  * In our case we have $Œ£Fy = 0$, and $Œ£œÑ_{A} = 0$ to solve for $F_A$ and $F_B$.


By following these steps, you'll be able to analyze and solve problems involving the forces and equilibrium of objects. These principles are fundamental in physics and engineering to understand how objects remain balanced and stationary.

## **Example**
**Given:**
- Ice Load Capacity = $P_{capacity}$ = $200 \text{lb}$
- Weight = W = 1500 lb

### **Thinking in steps:** ü§î
**Step 1**: Calculate Support Forces

To calculate the support forces, we can use the equations of equilibrium.

Equilibrium in the vertical direction:

$$
Œ£Fy = 0 = F_A + F_B - F_W
$$

Equilibrium in the horizontal direction (assuming no horizontal load):
$$
Œ£Fx = 0
$$

So far we have 2 unknowns and 2 equations but one of them (the horizontal one) doesnt actually give us any information. Last time when we looked at the person standing on the ice, we assumed that the weight was spread evenly between their two feet but here we don't have that assumption we can make.

This is where our torque equation comes into play!

$$
Œ£œÑ_A = 0 =  F_A \times (0 m) - 1500 lb \times (2 ft) + F_B (5 ft)
$$

Since the moment arm for $F_A$ is zero, that whole term drops out of the equation and becomes:

$$
Œ£œÑ_A = 0 =  - 1500 lb \times (2 ft) + F_B (5 ft)
$$

And now solving for $F_B$ we get

$$
\frac{1500 lb \times (2 ft)}{5 ft} = F_B
$$

$F_B$ = 600 lb

Now that we have $F_B$ we can substitute this back into our equation for vertical equilibrium to solve for $F_A$.

$$
Œ£Fy = 0 = F_A + F_B - F_W
$$

$$
Œ£Fy = 0 = F_A + 600 lb - 1500 lb
$$

$F_A$ = 900 lb

**Step 2:** Assess Stability

- If $F_A \leq P_{capacity}$ , the trailer is stable.
- If $F_A > P_{capacity}$ , the trailer is not stable.

**Results**

After performing the calculations, we obtain the following results:

* $F_A$ = 900 lb
* $F_B$ = 600 lb
* The trailer is not stable.

These results provide us with information about the support forces and the stability of the person based on the given data and calculations.


Disclaimer: It's important to note that while these hand calculations provide a simplified analysis, real-world scenarios can be more complex. Factors such as variations in ice strength, temperature, and the distribution of weight may impact the actual stability of the ice surface. Engineers and safety experts often employ safety factors and consider additional variables to ensure safety in practical situations. Therefore, these calculations serve as a starting point and should not be relied upon as the sole basis for assessing safety on ice. Always exercise caution and consult with experts when evaluating the stability of frozen surfaces in real-life situations. Safety should be the top priority in any such assessment.

```python
# This code cell can be deleted so the students have to write it on their own from scratch
# OR This code cell can be modified to be pseudo code or a code outline so students have to complete the code on their own
      # Here is an example of what you could give for a code outline
      # Set up your constants
      # Set FW to 1500
      # Set ice_load_capacity to 200
      # Set distance_FA to 2
      # Set distance_FB to 5
      # Step 1: Calculate Support Forces
      # Calculate FB using torque equation
      # Set FB to (FW multiplied by distance_FA) divided by distance_FB
      # Calculate FA using vertical equilibrium equation
      # Set FA to FW minus FB
      # Step 2: Assess Stability
      # If FA is less than or equal to ice_load_capacity display "The trailer is stable."
      # otherwise display "The trailer is not stable."
      # Display results
# OR This code cell can be left as is so students have an example to work from and they can be encouraged to just change values


# Constants
FW = 1500  # Weight of the truck in lb
ice_load_capacity = 200  # Ice bearing capacity in lb
distance_FA = 2  # Distance of the trailer's center of mass from leftmost tire in ft
distance_FB = 5  # Distance of point B from leftmost tire in ft

# Step 1: Calculate Support Forces

# Calculate FB using torque equation
FB = (FW * distance_FA) / distance_FB

# Calculate FA using vertical equilibrium equation
FA = FW - FB

# Step 2:  Assess Stability
if FA <= (ice_load_capacity):
    print("The trailer is stable.")
else:
    print("The trailer is not stable.")

# Display results
print("Support Force at A (FA):", FA, "lb")
print("Support Force at B (FB):", FB ,"lb")
```

This makes sense! We calculated this ice capacity for 4" and if you look back at the NOAA safety on ice infographic from the first lesson you will see that the minimum ice thickness guideline for a medium size truck is 12"! So let's calculate what our load capacity would be at 12" and see how we are doing then.

```python
# This code cell can be deleted so the students have to write it on their own from scratch
# OR This code cell can be modified to be pseudo code or a code outline so students have to complete the code on their own
      # Here is an example of what you could give for a code outline
      # Constants
      # Set A to 50
      # Set SF to 4
      # Set h to 12
      # Calculate Support Capacity
      # Calculate ice_load_capacity using the formula: ice_load_capacity = A * h^2 / SF
      # Display results
# This code cell can be left as is so students have an example to work from and they can be encouraged to just change values

# Constants
A = 50  # Value of A
SF = 4  # Safety factor
h = 12  # Given value of h in inches

# Calculate P using the formula
ice_load_capacity = A * h**2 / SF

# Print the calculated value of P
print("The ice_load_capacity is:", ice_load_capacity)
```
